```python
#L'installation d'anaconda vous fournira tout les bibliothque ncessaire pour la data Science
import numpy as np
import matplotlib.pyplot as plt

def box_muller(n):
    U1 = np.random.uniform(0, 1, n)
    U2 = np.random.uniform(0, 1, n)

    R = np.sqrt(-2 * np.log(np.where(U1 > 0, U1, 1e-10)))  # Replace 0 values with a small epsilon (1e-10)

    Theta = 2 * np.pi * U2
    Z = R * np.cos(Theta)

    return Z
```

L'algorithme de Box-Muller permet de gnrer des variables alatoires normalement distribues en utilisant les variables $U_1$ et $U_2$ comme suit :

\begin{align*}
R &= \sqrt{-2 \log(U_1)}, \\
\Theta &= 2\pi U_2, \\
Z &= R \cos(\Theta).
\end{align*}

La variable alatoire $Z$ obtenue ainsi suit une distribution normale standard.


Sachant que $U_1, U_2, \ldots$ une squence de variables alatoires indpendantes et identiquement distribues selon une loi normale standard.

<h5> 1.2 Les histogrammes des data massives gnres correspondant aux nombres n = 20, 50, 300, 1000 et 7000 </h5>

```python
def visualize_histograms(X,Y,n, alpha, var):
    X = box_muller(n)
    Y = alpha * box_muller(n) + np.sqrt(var) * np.random.randn(n)
    
    # Affichage de l'histogramme
    plt.hist(Y, bins='auto', alpha=alpha, color='blue')
    
    # Affichage du texte indiquant le nombre d'chantillons
    plt.text(0.5, -0.2, f"Histogrammes des donnes massives gnres correspondant  n = {n}",
             horizontalalignment='center', transform=plt.gca().transAxes)
    
    # Ajout des points sous forme de cercles vides
    plt.scatter(X, Y, marker='o', facecolors='none', edgecolors='red')
  
    plt.xlabel('X')
    plt.ylabel('Y')
    plt.title(f"Histogrammes des donnes massives gnres correspondant  n = {n}")
    
    plt.grid(True)
    plt.show()
```

```python
lst = [20,50,300,1000,7000]

alpha =0.7
var = 0.5

for n in lst:
    X = box_muller(n)
    Y = alpha * box_muller(n) + np.sqrt(var) * np.random.randn(n)
    visualize_histograms(X,Y,n, alpha, var)
```

<font color="green" ><h4>Intrprtation :</h4></font>

On Observe que plus on agglomre un grands nombre d'chantillons ensemble ,la distribution tend  ressembler  celle de la loi normal

<font color="green" ><h4>Demonstration Mathmatique:</h4></font>

\begin{align*}
X_1, X_2, \ldots, X_n & \text{ i.i.d. avec moyenne } \mu \text{ et variance } \sigma^2.\\
S_n &= X_1 + X_2 + \ldots + X_n.\\
E[S_n] &= n\mu, \quad \text{Var}(S_n) = n\sigma^2.\\
\text{Selon le thorme central limite,} \\
\lim_{{n \to \infty}} \frac{{S_n - n\mu}}{{\sqrt{n\sigma^2}}} &\approx \text{Distribution normale}.
\end{align*}

<font color="green" ><h4>Conclusion :</h4></font>

Cela dmontre que lorsque le nombre d'chantillons 

n devient trs grand, la somme des chantillons $S_n$
  tend  se rapprocher d'une distribution normale avec une moyenne de $\mu$
n et une variance de 2n

<h5>2.2 Estimation des valeurs, de la moyenne
$\mu$ et de lcart-type $\sigma$ pour X et Y </h5>

Comme X est un V.A qui suit la loi de Gauss a une fonction de densit  de  proba qui s'ecrit sous la forme
$$f_x(x)=\frac{1}{\sigma\sqrt{2\pi}}e^{-(\frac{x-\mu}{\sqrt{2}\sigma})^2}=K.E(x)$$

```python
import numpy as np
import matplotlib.pyplot as plt

def MMSE_viz_TCL(X):

    mean = np.mean(X)
    std = np.std(X)

    # Affichage de l'histogramme des donnes
    plt.hist(X, bins='auto', color='blue', alpha=0.7, density=True , label = 'Donne')

    x = np.linspace(np.min(X), np.max(X), 100)
    E = lambda x:np.exp(-0.5 * ((x - mean) / std) ** 2)
    K =  1 / (std * np.sqrt(2 * np.pi)) 
    y = K*E(x)

    # Affichage de la courbe de la loi normale
    plt.plot(x, y, color='red', label='Loi normale')

    plt.xlabel('Valeur des X')
    plt.ylabel('Frquence ou Probabilit')
    plt.title("Superposition de l'allure de l'histogramme des donnes et de la loi normale")
    plt.legend()
    plt.grid(True)
    plt.show()


MMSE_viz_TCL(X)
```

<font color="green" ><h4>Interprtation :</h4></font>

Les donnes massives gnres suivent bien la distribution de la loi normale centre rduite.

<font color="green" ><h4>Test de Normalit :</h4></font>

```python
from scipy import stats
statistic, p_value = stats.normaltest(X) 

# Afficher les rsultats
alpha = 0.05  # Niveau de signification
print("Rsultats du test de normalit:")
print(f"Statistique du test: {statistic}")
print(f"P-value: {p_value}")
if p_value < alpha:
    print("Les donnes ne suivent pas une distribution normale.")
else:
    print("Les donnes suivent une distribution normale.")
```

<h5>3.1 Estimateurs de  et  par la mthode de maximum de vraisemblance</h5>


Estimation par maximum de vraisemblance

Soit $(X_i \sim \text{Poisson}(\theta))$, o $(\theta)$ est le paramtre  estimer.
$$
X  \approx \epsilon(1/\mu) = \begin{cases} 0 & \text{si } x < 0 \\ x & \text{si } x \geq 0 \end{cases}
$$

$$
L_\theta(x_1,x_2,\dots,x_n) = \prod\limits_{i=1}^{n}{P(x=x_i)} = \frac{1}{\mu} e^{-\sum\limits_{i=1}^{n}x_i/n}
$$

Pour rsoudre $\theta$, on cherche :
$$
\widehat{\theta_{MV}} = \arg\max_{\theta} L(\theta)
$$
Ce qui revient  rsoudre :
$$
\frac{dL(\theta)}{d\theta} = 0 \quad \text{et} \quad \frac{d^2L(\theta)}{d\theta^2} < 0
$$
On obtient :
$$
\widehat{\theta_{MV}} = \frac{\sum\limits_{i=1}^{n}x_i}{n}
$$


Essayant d'estimer la valeur de $\mu$ et $\sigma$ avec leurs intervalles de confiances


$$
\text{Intervalle de confiance (95\%) pour la moyenne : } \bar{X} \pm z \left(\frac{\sigma}{\sqrt{n}}\right)
$$

$$
\text{Intervalle de confiance (95\%) pour l'cart-type : } s \pm z \left(\frac{s}{\sqrt{2(n-1)}}\right)
$$


```python
mean_MV, std_MV = stats.norm.fit(Y) # Similaire a la commandes

print("Estimation des paramtres par maximum de vraisemblance (MV):")
print("Moyenne MV:", mean_MV)
print("Intervalle de confiance (95%) de la moyenne MV:", stats.norm.interval(0.95, loc=mean_MV, scale=std_MV/np.sqrt(n)))
print("cart-type MV:", std_MV)
print("Intervalle de confiance (95%) de l'cart-type MV:", stats.norm.interval(0.95, loc=std_MV, scale=std_MV/np.sqrt(2*(n-1))))
```

```python
def MLE_viz_TCL(X):
    mean_MV, std_MV = stats.norm.fit(X)

    mean_formula = np.mean(X)
    std_formula = np.std(X)

    plt.hist(X, bins='auto', density=True, alpha=0.5, label='Donnes')
    x = np.linspace(np.min(X), np.max(X), 100)
    plt.plot(x, stats.norm.pdf(x, loc=mean_formula, scale=std_formula), 'r-', label='Loi normale (Formule)')
    plt.plot(x, stats.norm.pdf(x, loc=mean_MV, scale=std_MV), 'g-', label='Loi normale (MV)')
    plt.xlabel('Valeur des X')
    plt.ylabel('Frquence ou Probabilit')
    plt.title('Superposition des histogrammes et courbes de loi normale')
    plt.legend()
    plt.grid(True)
    plt.show()
    
MLE_viz_TCL(X)
```

```python
P = np.polyfit(X.flatten(), Y, 1) 
beta1,beta0 = P[0],P[1]
beta1,beta0
```

```python
plt.scatter(X, Y, color='b', label='Donnes massives') 
plt.plot(X, beta1*X + beta0, color='r', label='Droite de rgression') 
plt.title("Nuages de donnes massives avec ajustement linaire") 
plt.xlabel("X") 
plt.ylabel("Y") 
plt.legend() 
plt.show()
```

```python

print(f"La droite de rgression est : {beta1}x + {beta0}")
```

```python
#-----4.3------#
# Ajouter une colonne de 1  la matrice X
X = X.reshape(40000, 1)
X1 = np.concatenate((np.ones((len(X), 1)), X), axis=1)
estimateur = np.linalg.lstsq(X1, Y, rcond=None)[0]
b = estimateur[0] # Coefficient de l'ordonne  l'origine
a = estimateur[1] # Coefficient de la pente
print(f"Les estimateurs ponctuels sont a = {a} et b = {b}")
```

```python
#-----4.4------#
# Vrification si le point G appartient  la droite de rgression
Point_G = [np.mean(X), np.mean(Y)]
if np.isclose(Point_G[0] * beta1 + beta0, Point_G[1]):
    print("Le point G appartient  la droite de rgression")
else:
    print("Le point G n'appartient pas  la droite de rgression")
```

```python
#-----4.5------#
# Calcul du coefficient de dtermination (R)
corr = np.corrcoef(X.flatten(), Y.flatten())
R = np.sqrt((corr[0, 1] ** 2) / (np.var(X) * np.var(Y)))
print("Le coefficient de dtermination R =", R)
```

```python
#---6-4--- Calcul de la moyenne et de l'cart-type par les formules mathmatiques
mean_math = np.mean(data_ages)
std_math = np.std(data_ages)

# Estimation par maximum de vraisemblance (MLE)
mean_mle, std_mle = norm.fit(data_ages)

# Calcul de l'intervalle de confiance  95%
conf_int = norm.interval(0.95, loc=mean_mle, scale=std_mle)

# Affichage des rsultats
print("La moyenne calcule par formule mathmatique est :", mean_math)
print("L'cart-type calcul par formule mathmatique est :", std_math)
print("La moyenne calcule par MV est :", mean_mle)
print("L'cart-type calcul par MV est :", std_mle)
print("L'intervalle de confiance  95% :", conf_int)
```
